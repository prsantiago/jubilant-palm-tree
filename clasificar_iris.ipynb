```python
import pandas as pd
```

```python
iris = pd.read_csv(
    "datos/iris.data", header=None
)
```

```python
iris.shape
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# seleccionar setosa y versicolor
y = iris.iloc[:100, 4].values
y = np.where(y == 'Iris-setosa', -1, 1)

# extraer longitud de sépalo y longitud de pétalo
X = iris.iloc[:100, [0, 2]].values
```

```python
plt.scatter(X[:50, 0], X[:50, 1], color='red', marker='o', label='setosa')
plt.scatter(X[50:100, 0], X[50:100, 1], color='blue', marker='x', label='versicolor')

plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')
```

```python
# Entrenar el modelo de perceptrón
from clasificadores.perceptron import Perceptron
```

```python
ppn = Perceptron(eta=0.1, n_iter=10)
ppn.fit(X, y)
```

```python
plt.plot(range(1, ppn.n_iter+1), ppn.errors_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of errors')
plt.show()
```

```python
# Una función para representar regiones de decisión
from utils.visualizaciones import plot_decision_regions
```

```python
plot_decision_regions(X, y, ppn)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')

plt.show()
```
